```python
import matplotlib.pyplot as plt

def plot_value_array(i, predictions_array, true_label):
  true_label = true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('green')
  thisplot[true_label].set_color('blue')
```

```python
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt

fashion_mnist = tf.keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

img = test_images[2]

plt.imshow(img, cmap="gray")
```

```python
from google.cloud import aiplatform


img = (np.expand_dims(img,0)).tolist()
endpoint = aiplatform.Endpoint('projects/cloudapp-assigment-3/locations/us-east1/endpoints/1230560219667365888')

result = endpoint.predict(instances=img)
result.predictions
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plot_value_array(1, result.predictions[0], test_labels)
_ = plt.xticks(range(10), class_names, rotation=45)
plt.show()
```
